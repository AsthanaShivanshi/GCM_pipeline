```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
pr0 = pr.isel(time=5)

fig = plt.figure(figsize=(8, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
pr0.plot(ax=ax, transform=ccrs.PlateCarree(), cmap="Blues", cbar_kwargs={'label': 'Precipitation (mm/day)'})
ax.coastlines()
ax.set_extent([5.5, 10.5, 45.5, 48], crs=ccrs.PlateCarree()) 
plt.show()
```
